Question 1 : What are Type I and Type II errors in hypothesis testing, and how do they impact decision-making?

Answer : Type I and Type II errors are critical concepts in hypothesis testing, impacting how decisions are made based on data.

Definitions :

* Type I Error (False Positive):

This error occurs when the null hypothesis  is true, but is incorrectly rejected.​

It means detecting an effect or difference that does not actually exist.

The probability of making a Type I error is denoted by alpha (

α), often set at 0.05.​

* Type II Error (False Negative):

This happens when the null hypothesis is false, but fails to be rejected.​

It means failing to detect a real effect or difference.

The probability of making a Type II error is called beta (
β
β).​

* Impact on Decision-Making

1) Type I errors may cause one to take action or adopt changes based on an apparent effect that does not exist. For example, approving a new drug perceived as effective when it actually is not.​

2) Type II errors may lead to missed opportunities by failing to recognize a real effect. For example, not approving a genuinely effective treatment.​

Question 2:What is the P-value in hypothesis testing, and how should it be interpreted in the context of the null hypothesis?

Answer : In hypothesis testing, the P-value is a number that helps you decide whether to reject the null hypothesis. Specifically, the P-value measures the probability of getting results at least as extreme as the ones observed in your sample data, assuming that the null hypothesis is true.​

Here's how it should be interpreted: if the P-value is very small (commonly less than or equal to 0.05), it means that such an extreme outcome is unlikely under the null hypothesis. In this case, you have strong evidence against the null hypothesis, so you would reject it. However, if the P-value is greater than 0.05, the evidence against the null is weak, and you fail to reject the null hypothesis — meaning your data do not provide enough reason to say the null is false.

Question 3:Explain the difference between a Z-test and a T-test, including when to use each .

Answer : **Z-test (Use for Large Samples)**

Use a Z-test when you have a big sample (usually more than 30 people or items) and you already know the standard deviation of the entire population.​

* For example, suppose a company tracks the heights of all its employees every year. If they want to see if a new group of 50 hires fits the typical height, they would use a Z-test because they already know everyone's average height and variation.

The Z-test uses something called the normal distribution (which looks like a bell curve).​

**T-test (Use for Small Samples)**

Use a T-test when your sample is small (less than 30 people or items) or when you don’t know the population’s standard deviation.​

* For instance, a scientist tries a new medicine with just 12 patients and wants to see if it works better than the usual treatment. Because the group is so small and there's no info about the whole population, she’ll use a T-test.

The T-test uses the t-distribution, which is almost like the normal distribution but has fatter tails to work better with small samples.​

* Easy Rule

1) If you KNOW the population standard deviation or have a big group: use a Z-test.

2) If you DON’T KNOW the population standard deviation and your group is small: use a T-test.​

In real life, most people end up using the T-test unless they have a very big group or full info about the population. Both tests help us find out if there is a real difference between groups or if what we see could just be random chance.

Question 4:What is a confidence interval, and how does the margin of error influence its width and interpretation?

Answer : A confidence interval is a range that tells you where the true answer from a bigger group will likely fall, based on your sample data and a certain level of confidence—like 95%. Imagine you surveyed some people about their favorite fruit and found that, on average, most people like apples. Instead of saying, “The average is definitely 60%,” a confidence interval lets you say, “I’m pretty sure the real number falls between 55% and 65%.”​

The margin of error is what controls how wide this range is. If the margin of error is small, the confidence interval is narrow, which means the estimate is more precise.

Question 5: Describe the purpose and assumptions of an ANOVA test. How does it extend hypothesis testing to more than two groups?

Answer : An ANOVA (Analysis of Variance) test is used to find out if there are real differences between the averages (means) of three or more groups, instead of just two. Researchers use ANOVA when they want to see if changing a factor (like a treatment, type of food, or teaching method) affects different groups in their study.​

* Purpose of ANOVA

The main purpose is to see if at least one group's average is different from the others. Instead of running many two-group tests (which can increase errors), ANOVA allows you to check all groups together in one go, giving a clearer picture while avoiding mistakes from too many comparisons. If the ANOVA test finds a significant difference, it tells you that not all groups are the same, but you'll need extra steps to find out which one is different.​

* Assumptions of ANOVA

For the results to be trustworthy, three big assumptions should be met:​

1) Independence: Each group's results are independent—they don't affect each other.​

2) Normality: The data in every group should follow a normal (bell-shaped) curve, with most values near the average.​

3) Homogeneity of Variance: The amount of spread (variance) in scores should be roughly the same in each group.​

4) Random Sampling: The samples should be chosen randomly so they represent their groups well.​

* Extending Hypothesis Testing to Multiple Groups :  

A t-test only compares two groups (for example, treatment vs. control), but ANOVA allows you to compare three, four, or even more at once. This is helpful in studies where you want to see if there are differences among several groups (like three different diets), without having to do many separate tests

Question 6: Write a Python program to perform a one-sample Z-test and interpret the result for a given dataset.

Answer : Python Program for One-Sample Z-Test
Suppose we know the average IQ in a population is 100, with a standard deviation of 15. After giving a treatment to 20 people, we get their test scores. We want to check if the treatment changes the average IQ.


```python
import numpy as np
from statsmodels.stats.weightstats import ztest

# Sample data (IQ scores after treatment for 20 people)
data = [88, 92, 94, 94, 96, 97, 97, 97, 99, 99, 100, 101, 103, 105, 108, 110, 111, 112, 114, 115]

# Perform one-sample Z-test, assuming population mean is 100
z_score, p_value = ztest(data, value=100)

print("Z-score:", z_score)
print("P-value:", p_value)
```

*Explanation in Simple Language

1) The Z-score tells us how far our sample mean is from the population mean, in terms of standard deviations.

2) The p-value tells us how likely it is to get a sample mean at least as extreme as ours if the population mean is really 100.

3) If the p-value is less than 0.05 (for a 5% significance level), it means the difference is statistically significant. This means our sample data suggests the average IQ really changed after treatment. If the p-value is larger, it means there is no strong evidence of a change.

* All you need to remember:

1) Low p-value (< 0.05): The sample mean is probably different from the population mean. Something is likely going on!

2) High p-value (> 0.05): There’s no strong reason to think anything changed. The sample mean is not very different from what’s expected.​

Question 7:Simulate a dataset from a binomial distribution (n = 10, p = 0.5) using NumPy and plot the histogram.

Question 8: Generate multiple samples from a non-normal distribution and implement the Central Limit Theorem using Python.

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate a population from a non-normal distribution (Gamma)
population_size = 10000
shape_param = 2.0  # shape parameter for gamma distribution
population = np.random.gamma(shape=shape_param, scale=1.0, size=population_size)

# Step 2: Specify sample size and number of samples
sample_size = 30  # size of each sample
num_samples = 1000  # number of samples to draw

# Step 3: Draw samples and compute their means
sample_means = [np.mean(np.random.choice(population, sample_size)) for _ in range(num_samples)]

# Step 4: Plot the distribution of the original population (non-normal)
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(population, bins=30, edgecolor='black', alpha=0.7)
plt.title('Original Population Distribution (Gamma)')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Step 5: Plot the distribution of the sample means
plt.subplot(1, 2, 2)
plt.hist(sample_means, bins=30, edgecolor='black', alpha=0.7)
plt.title('Distribution of Sample Means (CLT Effect)')
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

Question 9: Write a Python function to calculate and visualize the confidence interval for a sample mean.

```python
import numpy as np
import scipy.stats as stats

data = [12, 15, 14, 10, 13, 17, 14, 15, 16, 14]

mean = np.mean(data)
sem = stats.sem(data)  # Standard error of the mean
confidence = 0.95
lower, upper = stats.t.interval(confidence, len(data)-1, loc=mean, scale=sem)

print(f"Mean: {mean:.2f}")
print(f"95% confidence interval: ({lower:.2f}, {upper:.2f})")
```

Question 10: Perform a Chi-square goodness-of-fit test using Python to compare  observed and expected distributions, and explain the outcome.

```python
import numpy as np
from scipy.stats import chisquare

# Example data: observed frequencies
observed = np.array([18, 22, 20, 15, 25])

# Expected frequencies (if all categories are equally likely)
expected = np.array([20, 20, 20, 20, 20])

# Perform Chi-square goodness-of-fit test
chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

print(f"Chi-square statistic: {chi2_stat:.2f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
if p_value < 0.05:
    print("Reject the null hypothesis: The observed distribution differs significantly from the expected.")
else:
    print("Fail to reject the null hypothesis: No significant difference between observed and expected distributions.")
```
