  9. Write a python program to convert user input from string to integer ,float and Boolean.

```python
# Python program to convert user input into int, float, and bool

# Ask the user for input (it always comes as a string)
user_input = input("Enter something: ")


# Convert to integer
try:
    int_value = int(user_input)
    print(f"As Integer: {int_value}   (Example: '25' â†’ 25)")
except ValueError:
    print("As Integer: Cannot convert   (Example: 'hello' âŒ)")

# Convert to float
try:
    float_value = float(user_input)
    print(f"As Float: {float_value}   (Example: '3.14' â†’ 3.14)")
except ValueError:
    print("As Float: Cannot convert   (Example: 'world' âŒ)")

# Convert to Boolean
# Rule: empty string "" â†’ False, anything else â†’ True
bool_value = bool(user_input)
print(f"As Boolean: {bool_value}   (Example: '' â†’ False, 'yes' â†’ True)")
```

  10. Write code to demonstrate type casting with list elements.

```python
# Original list with strings
my_list = ["10", "20", "30", "40.5", ""]

print("Original list (strings):", my_list)

# Convert list elements to integers
int_list = []
for item in my_list:
    try:
        int_list.append(int(float(item)))  # convert to float first for decimal strings
    except ValueError:
        int_list.append(None)  # if conversion fails
print("Converted to integers:", int_list)

# Convert list elements to floats
float_list = []
for item in my_list:
    try:
        float_list.append(float(item))
    except ValueError:
        float_list.append(None)
print("Converted to floats:", float_list)

# Convert list elements to booleans
bool_list = [bool(item) for item in my_list]  # empty string -> False, non-empty -> True
print("Converted to booleans:", bool_list)
```

13. Write a Python program to find the sum of all even numbers between 1 to 50 .

```python
# Initialize sum
total = 0

# Loop through numbers from 1 to 50
for num in range(1, 51):
    if num % 2 == 0:  # Check if the number is even
        total += num  # Add the even number to the total

# Print the result
print("The sum of all even numbers from 1 to 50 is:", total)
```

 14. Write a program to reverse a string using a while loop.  

```python
# Ask the user to enter a string
string = input("Enter a string: ")

# Initialize variables
reversed_string = ""  # This will store the reversed string
index = len(string) - 1  # Start from the last character

# Use while loop to reverse the string
while index >= 0:
    reversed_string += string[index]  # Add the character at current index
    index -= 1  # Move to the previous character

# Print the reversed string
print("Reversed string:", reversed_string)
```

 15. Write a Python program to calculate the factorial of a number provided by the user using a while loop.

```python
num = int(input("Enter a number: "))
factorial = 1
i = 1

while i <= num:
  factorial *= i
  i += 1

  print("Factorial of", num, "is:", factorial)
```
