6. Explain the concept of generators in Python and how they are defined.

Ans. üîπ  Generator in Python:

A generator is like a function that remembers where it left off.

Instead of giving you all results at once (like a list), it gives you results one at a time, when you ask for them.

This makes generators memory-efficient ‚Äî useful when you‚Äôre dealing with large data or infinite sequences.

üîπ How do we define a generator?

There are two ways:

1)Using yield inside a function

Normally, functions use return (which ends the function).

But with yield, the function pauses and can continue later.

2)Using generator expressions

Similar to list comprehensions, but with () instead of [].

üîπ Example : Generator function with yield

def count_up_to(n):

    num = 1
    while num <= n:
        yield num  # pauses and returns the current number
        num += 1

 #Using the generator

for number in count_up_to(5):

    print(number)

7. What are the advantages of using generators over regular functions?

Ans. üîπ Advantages of generators over regular functions

1)Save memory

They don‚Äôt store the whole result in memory.

2)Faster for big data

3)Pause and resume

A generator ‚Äúremembers‚Äù where it stopped and continues from there.

1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.




```python
def sum_of_evens(numbers):
    total = 0
    for num in numbers:
        if num % 2 == 0:   # check if the number is even
            total += num   # add it to total
    return total

my_list = [1, 2, 3, 4, 5, 6]

# Call the function
result = sum_of_evens(my_list)

print("The sum of even numbers is:", result)
```

4. Write a Python function that checks if a given number is prime or not from 1 to 200.


```python
# Function to check if a number is prime
def is_prime(n):
    if n <= 1:
        return False  # 0 and 1 are not prime numbers
    for i in range(2, int(n**0.5) + 1):  # Check from 2 to sqrt(n)
        if n % i == 0:
            return False
    return True

# Print prime numbers from 1 to 200
for number in range(1, 201):
    if is_prime(number):
        print(number, end=" ")
```

5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.


```python
class fibonacciIterator:
  def __init__(self, n):
    self.n= n
    self.a, self.b, self.count = 0, 1, 0

  def __iter__(self):
    return self

  def __next__(self):
    if self.count < self.n:
       result = self.a
       self.a, self.b = self.b, self.a + self.b
       self.count += 1
       return result
    else:
      raise StopIteration

# Example usage
for number in fibonacciIterator(10):
    print(number)
```

6. Write a generator function in Python that yields the powers of 2 up to a given exponent.


```python
# Generator function
def powers_of_two(max_exp):
    for i in range(max_exp + 1):  # Go from 0 to max_exp
        yield 2 ** i  # Yield 2 to the power of i

# Example usage
for value in powers_of_two(5):
    print(value)
```

7. Implement a generator function that reads a file line by line and yields each line as a string.

```python
# Create a dummy file named myfile.txt for demonstration
with open("myfile.txt", "w") as f:
    f.write("This is the first line.\n")
    f.write("This is the second line.\n")
    f.write("And this is the third line.")

    # Generator function
def read_file_line_by_line(filename):
    with open(filename, "r") as file:
        for line in file:
            yield line.strip()  # yield each line without extra \n

# Example usage:
for line in read_file_line_by_line("myfile.txt"):
    print(line)
```

9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

```python
# List of temperatures in Celsius
celsius = [0, 10, 20, 30, 40]

# Convert Celsius to Fahrenheit using map()
fahrenheit = list(map(lambda c: (c * 9/5) + 32, celsius))

# Print results
print("Celsius:    ", celsius)
print("Fahrenheit: ", fahrenheit)
```

10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

```python
# Program to remove vowels using filter()

def remove_vowels(s):
    vowels = "aeiouAEIOU"
    return "".join(filter(lambda ch: ch not in vowels, s))

# Example
text = "Hello World"
result = remove_vowels(text)
print("Original:", text)
print("Without vowels:", result)
```
